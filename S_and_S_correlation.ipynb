```python
"""
Since there are 106 columns, I'm checking for correlations between them. 
Correlation in simple terms means close linear relationship between two features so that one feature can be dropped. 
I'm dropping all the features having correlation greater than equal to 0.9.
"""
correlation = df.corr()
```

```python
plt.figure(figsize =(25,25))
sns.heatmap(df[correlation.index].corr(),cmap="RdYlGn",linewidths=1.0)
```

```python
columns = np.full((correlation.shape[0],), True, dtype=bool)
#print(columns_copy.shape)
for i in range(correlation.shape[0]):
    for j in range(i+1, correlation.shape[0]):
        if correlation.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
#print(columns.shape)

columns_to_consider = df.columns[columns]

df = df[columns_to_consider]
#print(len(df))"""
```

### Step 7: Creating Dataset for Model Training

```python
"""
Here I'm considering this as a Classification task. There are 36 unique classess in "Amazon Categories". So this becomes my 
ground truth/ labels/ dependent varible. Rest of the 57 columns are the independent varibales.
"""
y = df.iloc[:,35].values
X = df.drop('amazoncategories',axis=1,inplace=True)
X = df.iloc[:,:].values
print("Shape of X is:", X.shape)
print("Shape of y is:", y.shape)
```

```python
"""df_majority = df[df["Attrition"]==0]
df_minority = df[df["Attrition"]==1]
 
# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=3699,    # to match majority class
                                 random_state=0) # reproducible results
 
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])

yes = df_upsampled[df_upsampled["Attrition"]==1]
no = df[df["Attrition"]==0]

print("Total number of examples are \n", len(df_upsampled))
print("Number of yes:",len(yes), "\t Number of no:", len(no), "\t Total is:",len(yes+no))

sns.countplot(x = "Attrition",data=df_upsampled)
plt.show()"""

from imblearn.over_sampling import RandomOverSampler

rm = RandomOverSampler(sampling_strategy='auto',random_state=12)
X_res, y_res = rm.fit_sample(X, y)
plt.figure(figsize =(20,20))
sns.countplot(y_res)
#x_val_res, y_val_res = sm.fit_sample(X_val, y_val)
```

```python
print("Shape of X is:", X_res.shape)
print("Shape of y is:", y_res.shape)
```

```python
def Random_forest_model(X_train, y_train,X_val,y_val):
    """
    This function uses Random Forest trees for classification
    
    It returns training and validation accurcy of each fold
    """
    
    rf_c = RandomForestClassifier()
    rf_c.fit(X_train,y_train)
    score_train_acc = rf_c.score(X_train, y_train)
    print("**** TRAINING INFORMATION ****")
    print("Training Accuracy: ", score_train_acc)
    score_val_acc = rf_c.score(X_val, y_val)
    print("**** Validation INFORMATION ****")
    print("Validation Accuracy: ", score_val_acc) 
    print("------------------------------")
    print("------------------------------")
    print("**** Metrics and Scoring ****")
    y_pred_RF = rf_c.predict(X_val)
    print(classification_report(y_val, y_pred_RF)) # Precision, Recall, F1 score, Accuracy
    print(confusion_matrix(y_val, y_pred_RF)) 
    # Saving Trained model
    pkl_filename = "Random_Forest_pickle_model.pkl"
    with open(pkl_filename, 'wb') as file:
        pickle.dump(rf_c, file)
    
    return score_train_acc, score_val_acc
```

```python
train_acc, val_acc = [], [] # Lists for storing accuracies on each fold
n_folds = 5 # 5 fold cross-validation

# Cross validation
kfold = KFold(n_splits=n_folds, shuffle=True, random_state=7)

for train, val in kfold.split(X_res, y_res):
    
    X_train, X_val = X_res[train], X_res[val] # Grabbing X_train, X_val for Model
    y_train, y_val = y_res[train], y_res[val] # Grabbing y_train, y_val for Models
    
    
    rf_model = Random_forest_model(X_train, y_train,X_val,y_val)
    #lr_model = LR_model(X_train, y_train,X_val,y_val)
    
    train_acc.append(rf_model[0])
    val_acc.append(rf_model[1])
    
    print("Training Accuracy is:", sum(train_acc)/n_folds) # Considering Training Mean Accuracy of every fold
    print("Validation Accuracy is:", sum(val_acc)/n_folds) # Considering Validation Mean Accuracy of every fold
```
